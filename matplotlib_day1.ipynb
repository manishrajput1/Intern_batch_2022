```python
angles=np.arange(0,361,15)
```

```python
angles
```

```python
rad=np.deg2rad(angles)
```

```python
rad
```

```python
sin=np.sin(rad)
cos=np.cos(rad)
```

```python
sin
```

```python
plt.figure(figsize=(8,4))
plt.plot(angles,sin,c='r',lw=3,ls=':',label='sin')
plt.plot(angles,cos,c='g',lw=3,ls=':',label='cos')

plt.grid()
plt.legend(loc=4)
plt.xticks([0,20,40,60,80,100,120,140,160,180,200,220,240,260,280,300,320,340,360])
plt.xlabel('Angles',fontsize=15,color='g')
plt.ylabel('Sin & cos values',fontsize=15,color='g')
plt.title('Sin & Cos plot',fontsize=15,color='g')
plt.savefig('line_plot_.png')
plt.show()
```
